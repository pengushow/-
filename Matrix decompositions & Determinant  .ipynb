```python
import numpy as np
```

```python
#法一(消去)



def ryDet1(A): 

     

    n= A.shape[0] 

    if n>1: 

        i= 0 

        Bi=  np.delete(A, i, axis=0) 

        #print(f'\ni= {i}, Bi= \n{Bi}') 

         

        det= 0 

        for j in range(n): 

             

            Bij= np.delete(Bi,j, axis=1) 

            print(f'\ni,j= {i}{j}, Bij= \n{Bij}') 

             

            Dij= ryDet1(Bij) #np.linalg.det(Bij) 

            print(f'.... Dij= {Dij}') 

             

            Cij= A[i,j]*Dij*(-1)**(i+j) 

            det += Cij 

        #print(f'det= {det}') 

        return det 

     

    elif A.shape[0]==1: 

        det= A[0,0] 

        return det 

    else: 

        return None 

 
```

```python
#法二(組合)
def ryDet2(A):
    det2=0
    m=A.shape[0]
    print("det=")
    for n in range(m):
        AA=[[0]*(m-1)for i in range(m-1)]
        for i in range(m-1):
            for j in range(m-1):
                if (j>=n):
                    AA[i][j]=A[i+1][j+1]
                else:
                    AA[i][j]=A[i+1][j]  
        print(f'\n{A[0][n]*((-1)**n)}*{np.array(AA)}') 
        det2+=A[0][n]*((-1)**n)*np.linalg.det(AA) 
    print(f'det= {round(det2)}') 
```

```python
#生成矩陣
n= 5  

 

np.random.seed(2) 

A= np.random.randint(-5,5,size=(n,n)) 

print(f'A= \n{A}')  

det= np.linalg.det(A) 

print(f'det= {round(det)}') 
```
